```python
import math
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

from sklearn import preprocessing
```

```python
plt.style.use('ggplot')
```

```python
def scaleFeatures(df):
    # Feature scaling is a type of transformation that only changes the
    # scale, but not number of features. Because of this, we can still
    # use the original dataset's column names... so long as we keep in
    # mind that the _units_ have been altered:

    scaled = preprocessing.StandardScaler().fit_transform(df)
    scaled = pd.DataFrame(scaled, columns=df.columns)
    
    print("New Variances:\n", scaled.var())
    print("New Describe:\n", scaled.describe())
    return scaled
```

```python
def drawVectors(transformed_features, components_, columns, plt, scaled):
    if not scaled:
        return plt.axes() # No cheating ;-)

    num_columns = len(columns)

    # This funtion will project your *original* feature (columns)
    # onto your principal component feature-space, so that you can
    # visualize how "important" each one was in the
    # multi-dimensional scaling

    # Scale the principal components by the max value in
    # the transformed set belonging to that component
    xvector = components_[0] * max(transformed_features[:,0])
    yvector = components_[1] * max(transformed_features[:,1])

    ## visualize projections

    # Sort each column by it's length. These are your *original*
    # columns, not the principal components.
    important_features = { columns[i] : math.sqrt(xvector[i]**2 + yvector[i]**2) for i in range(num_columns) }
    important_features = sorted(zip(important_features.values(), important_features.keys()), reverse=True)
    print("Features by importance:\n", important_features)

    ax = plt.axes()

    for i in range(num_columns):
        # Use an arrow to project each original feature as a
        # labeled vector on your principal component axes
        plt.arrow(0, 0, xvector[i], yvector[i], color='b', width=0.0005, head_width=0.02, alpha=0.75)
        plt.text(xvector[i]*1.2, yvector[i]*1.2, list(columns)[i], color='b', alpha=0.75)

    return ax
```

```python
scaleFeatures = True
```

```python
df = pd.read_csv("Datasets/kidney_disease.csv", sep=",", index_col=False)
df = df.dropna()
print (df)
```

```python
df_non_nominal = df.drop(['id', 'classification', 'rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane'], axis=1)
print (df_non_nominal)
```

```python
labels = ['red' if i=='ckd' else 'green' for i in df.classification]
```

```python
print (df_non_nominal.dtypes)
df_non_nominal['pcv'] = pd.to_numeric(df_non_nominal["pcv"], errors='coerce')
df_non_nominal['wc'] = pd.to_numeric(df_non_nominal["wc"], errors='coerce')
df_non_nominal['rc'] = pd.to_numeric(df_non_nominal["rc"], errors='coerce')
print (df_non_nominal.dtypes)
```

```python
print (df_non_nominal.var(axis=0))
df_non_nominal.describe()
```

```python
df = df_non_nominal
if scaleFeatures: df = scaleFeatures(df)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2, svd_solver="full")
pca.fit(df)
T = pca.transform(df)
```

```python
# Since we transformed via PCA, we no longer have column names; but we know we
# are in `principal-component` space, so we'll just define the coordinates accordingly:

ax = drawVectors(T, pca.components_, df.columns.values, plt, scaleFeatures)
T  = pd.DataFrame(T)

T.columns = ['component1', 'component2']
T.plot.scatter(x='component1', y='component2', marker='o', c=labels, alpha=0.75, ax=ax)

plt.show()
```

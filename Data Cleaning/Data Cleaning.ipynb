Import and alias Pandas:

```python
import pandas as pd
```

Load up the table from the link, and extract the dataset out of it. If you're having issues with this, look carefully at the sample code provided in the reading:

```python
table_dataframe = pd.read_html('http://www.espn.com/nhl/statistics/player/_/stat/points/sort/points/year/2015/seasontype/2')[0]
```

Next up, rename the columns so that they are _similar_ to the column definitions provided to you on the website. Be careful and don't accidentally use any column names twice. If a column uses special characters, you can replace them with regular characters to make it easier to work with:

```python
table_dataframe = table_dataframe.drop(table_dataframe.index[0])
#print (table_dataframe)
table_dataframe= table_dataframe.rename(columns=table_dataframe.iloc[0])
#print (table_dataframe)
table_dataframe = table_dataframe.drop(table_dataframe.index[0])
table_dataframe = table_dataframe.reset_index(drop = True)
table_dataframe=table_dataframe.rename(columns = {'+/-':'PlusMinusRating'})
table_dataframe=table_dataframe.rename(columns = {'PTS/G':'PointsperGame'})
table_dataframe.columns.values[13] = "PowerPlayGoals"
table_dataframe.columns.values[14] = "PowerPlayAssists"
table_dataframe.columns.values[15] = "ShortHandedGoals"
table_dataframe.columns.values[16] = "ShortHandedAssists"
print (table_dataframe)
```

Get rid of any row that has at least 4 NANs in it. That is, any rows that do not contain player points statistics:

```python
table_dataframe = table_dataframe.dropna(axis=0, thresh=4)
```

At this point, look through your dataset by printing it. There probably still are some erroneous rows in there. What indexing command(s) will you use to select all rows EXCEPT those rows?

```python
print (table_dataframe)
repeated_index_df = table_dataframe.index.isin([11,23,35])
table_dataframe = table_dataframe[~repeated_index_df]
print (table_dataframe)
```

Get rid of the 'RK' column:

```python
table_dataframe = table_dataframe.drop('RK', axis=1)
print (table_dataframe)
```

Make sure there are no holes in your index by resetting it. There is an example of this in the reading material. By the way, drop the original index.

```python
table_dataframe = table_dataframe.reset_index(drop = True)
print (table_dataframe)
```

Check the data type of all columns, and ensure those that should be numeric are numeric.

```python
print (table_dataframe.dtypes)
table_dataframe.GP = pd.to_numeric(table_dataframe.GP, errors='coerce')
table_dataframe.G = pd.to_numeric(table_dataframe.G, errors='coerce')
table_dataframe.A = pd.to_numeric(table_dataframe.A, errors='coerce')
table_dataframe.PTS = pd.to_numeric(table_dataframe.PTS, errors='coerce')
table_dataframe.PlusMinusRating = pd.to_numeric(table_dataframe.PlusMinusRating, errors='coerce')
table_dataframe.PIM = pd.to_numeric(table_dataframe.PIM, errors='coerce')
table_dataframe.PointsperGame = pd.to_numeric(table_dataframe.PointsperGame, errors='coerce')
table_dataframe.SOG = pd.to_numeric(table_dataframe.SOG, errors='coerce')
table_dataframe.PCT = pd.to_numeric(table_dataframe.PCT, errors='coerce')
table_dataframe.GWG = pd.to_numeric(table_dataframe.GWG, errors='coerce')
table_dataframe.PowerPlayGoals = pd.to_numeric(table_dataframe.PowerPlayGoals, errors='coerce')
table_dataframe.PowerPlayAssists = pd.to_numeric(table_dataframe.PowerPlayAssists, errors='coerce')
table_dataframe.ShortHandedGoals = pd.to_numeric(table_dataframe.ShortHandedGoals, errors='coerce')
table_dataframe.ShortHandedAssists = pd.to_numeric(table_dataframe.ShortHandedAssists, errors='coerce')
print (table_dataframe.dtypes)
```
